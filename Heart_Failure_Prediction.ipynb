# **K - MEANS WITH DIFFERENT DISTANCES**

K-MEANS ( WITH MINKOWSKI DISTANCE )

---



```python
import numpy as np
from scipy.spatial.distance import minkowski
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score
from scipy.stats import mode
import pandas as pd
data = pd.read_csv("/content/heart.csv")
for column in data.select_dtypes(include='object').columns:
    data[column] = LabelEncoder().fit_transform(data[column])
X = data.drop("HeartDisease", axis=1).values
y = data["HeartDisease"].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
def kmeans_minkowski(X, n_clusters=3, p=1.5, max_iter=100, tol=1e-4):
    np.random.seed(0)
    initial_centroids_idx = np.random.choice(len(X), n_clusters, replace=False)
    centroids = X[initial_centroids_idx]
    for _ in range(max_iter):
        labels = np.array([np.argmin([minkowski(x, centroid, p=p) for centroid in centroids]) for x in X])
        new_centroids = np.array([X[labels == j].mean(axis=0) for j in range(n_clusters)])
        if np.all(np.abs(new_centroids - centroids) < tol):
            break
        centroids = new_centroids
    return labels, centroids
n_clusters = len(np.unique(y))
labels, centroids = kmeans_minkowski(X_scaled, n_clusters=n_clusters, p=1.5)
label_mapping = np.zeros_like(labels)
for cluster in range(n_clusters):
    mask = (labels == cluster)
    label_mapping[mask] = mode(y[mask])[0]
accuracy = accuracy_score(y, label_mapping)
print("Clustering accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y, label_mapping))
print("Confusion Matrix:")
print(confusion_matrix(y, label_mapping))
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from scipy.optimize import linear_sum_assignment
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
data = pd.read_csv('/content/heart.csv')
for column in data.select_dtypes(include='object').columns:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
X = data.drop(columns=['HeartDisease'])
y_true = data['HeartDisease']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
n_clusters = len(np.unique(y_true))
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
y_pred = kmeans.fit_predict(X_scaled)
def calculate_accuracy(y_true, y_pred):
    contingency_matrix = np.zeros((len(set(y_true)), len(set(y_pred))), dtype=int)
    for i in range(len(y_true)):
        contingency_matrix[y_true.iloc[i], y_pred[i]] += 1
    row_ind, col_ind = linear_sum_assignment(-contingency_matrix)
    accuracy = contingency_matrix[row_ind, col_ind].sum() / len(y_true)
    return accuracy
accuracy = calculate_accuracy(y_true, y_pred)
print("Clustering Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y, label_mapping))
print("Confusion Matrix:")
print(confusion_matrix(y, label_mapping))
```

K - MEANS ( WITH MAHALANOBIS DISTANCE )

---



```python
import numpy as np
from scipy.spatial.distance import mahalanobis
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score
from scipy.stats import mode
import pandas as pd
data = pd.read_csv("/content/heart.csv")
for column in data.select_dtypes(include='object').columns:
    data[column] = LabelEncoder().fit_transform(data[column])
X = data.drop("HeartDisease", axis=1).values
y = data["HeartDisease"].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
cov_matrix = np.cov(X_scaled, rowvar=False)
inv_cov_matrix = np.linalg.inv(cov_matrix)
def kmeans_mahalanobis(X, inv_cov_matrix, n_clusters=3, max_iter=100, tol=1e-4):
    np.random.seed(0)
    initial_centroids_idx = np.random.choice(len(X), n_clusters, replace=False)
    centroids = X[initial_centroids_idx]
    for _ in range(max_iter):
        labels = np.array([np.argmin([mahalanobis(x, centroid, inv_cov_matrix) for centroid in centroids]) for x in X])
        new_centroids = np.array([X[labels == j].mean(axis=0) for j in range(n_clusters)])
        if np.all(np.abs(new_centroids - centroids) < tol):
            break
        centroids = new_centroids
    return labels, centroids
n_clusters = len(np.unique(y))
labels, centroids = kmeans_mahalanobis(X_scaled, inv_cov_matrix, n_clusters=n_clusters)
label_mapping = np.zeros_like(labels)
for cluster in range(n_clusters):
    mask = (labels == cluster)
    label_mapping[mask] = mode(y[mask])[0]
accuracy = accuracy_score(y, label_mapping)
print("Clustering accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y, label_mapping))
print("Confusion Matrix:")
print(confusion_matrix(y, label_mapping))
```

K - MEANS ( WITH JACCARD SCORE )

---



```python
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, jaccard_score, classification_report, confusion_matrix
from scipy.stats import mode
import pandas as pd
data = pd.read_csv("/content/heart.csv")
for column in data.select_dtypes(include='object').columns:
    data[column] = LabelEncoder().fit_transform(data[column])
X = data.drop("HeartDisease", axis=1).values
y = data["HeartDisease"].values
def jaccard_distance(a, b):
    intersection = np.sum(np.minimum(a, b))
    union = np.sum(np.maximum(a, b))
    return 1 - intersection / union
def kmeans_jaccard(X, n_clusters=3, max_iter=100, tol=1e-4):
    np.random.seed(0)
    initial_centroids_idx = np.random.choice(len(X), n_clusters, replace=False)
    centroids = X[initial_centroids_idx]
    for _ in range(max_iter):
        labels = np.array([np.argmin([jaccard_distance(x, centroid) for centroid in centroids]) for x in X])
        new_centroids = np.array([X[labels == j].mean(axis=0) for j in range(n_clusters)])
        if np.all(np.abs(new_centroids - centroids) < tol):
            break
        centroids = new_centroids
    return labels, centroids
n_clusters = len(np.unique(y))
labels, centroids = kmeans_jaccard(X, n_clusters=n_clusters)
label_mapping = np.zeros_like(labels)
for cluster in range(n_clusters):
    mask = (labels == cluster)
    label_mapping[mask] = mode(y[mask])[0]
accuracy = accuracy_score(y, label_mapping)
print("Clustering accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y, label_mapping))
print("Confusion Matrix:")
print(confusion_matrix(y, label_mapping))
```

#**RANDOM FOREST CLASSIFIER**


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
data = pd.read_csv("/content/heart.csv")
label_encoder = LabelEncoder()
for column in data.select_dtypes(include=['object']).columns:
    data[column] = label_encoder.fit_transform(data[column])
X = data.drop(columns=["HeartDisease"])
y = data["HeartDisease"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
